```python
from matplotlib import pyplot as plt
import numpy as np
import cv2

import model_fun
```

```python
def PSNR(original, denoised):
    mse = np.mean((original - denoised) ** 2)
    if(mse == 0):  
        return 100
    max_pixel = 255.0
    psnr = 20 * np.log10(max_pixel / np.sqrt(mse))
    return psnr
```

```python
def get_results(img, sigma, mu=0):
    noise = np.random.normal(mu, sigma, img.shape)
    noise = noise.astype(np.uint8)
    img_noisy = cv2.addWeighted(img, 0.5, noise, 0.5, 0)
    # Denoising 
    dst = np.zeros_like(img_noisy)
    dst = model_fun(img_noisy)

    return img_noisy, dst
```

```python
img1 = cv2.imread('test_img.jpeg',0)

noisy_img1, denoised_img1 = get_results(img1, sigma=25)
noisy_img2, denoised_img2 = get_results(img1, sigma=50)
noisy_img3, denoised_img3 = get_results(img1, sigma=75)

plt.figure(figsize=(20,15))
plt.subplot(331),plt.imshow(img1, cmap="gray")
plt.title("Original Image")
plt.subplot(332),plt.imshow(noisy_img1, cmap="gray")
plt.title("Noisy Image (Sigma = 25)")
plt.subplot(333),plt.imshow(denoised_img1, cmap="gray")
plt.title(f"Denoised Image (PSNR = {PSNR(noisy_img1, denoised_img1)})")
plt.subplot(334),plt.imshow(img1, cmap="gray")
plt.title("Original Image")
plt.subplot(335),plt.imshow(noisy_img2, cmap="gray")
plt.title("Noisy Image (Sigma = 50)")
plt.subplot(336),plt.imshow(denoised_img2, cmap="gray")
plt.title(f"Denoised Image (PSNR = {PSNR(noisy_img2, denoised_img2)})")
plt.subplot(337),plt.imshow(img1, cmap="gray")
plt.title("Original Image")
plt.subplot(338),plt.imshow(noisy_img3, cmap="gray")
plt.title("Noisy Image (Sigma = 75)")
plt.subplot(339),plt.imshow(denoised_img3, cmap="gray")
plt.title(f"Denoised Image (PSNR = {PSNR(noisy_img3, denoised_img3)})")
plt.tight_layout()
plt.savefig('result001_denoising.png')
plt.show()
```
